```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Perform standardization on the float columns
float_columns = ['ld','m0', 'm1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10', 'm11', 'm12', 'm13', 'm14']
scaler = StandardScaler()
train_data[float_columns] = scaler.fit_transform(train_data[float_columns])
```

```python
train_data
```

```python
# Replace missing values in each column
missing_counts = {'ld': 222, 'm3': 314, 'm4': 467, 'm5': 576, 'm6': 675, 'm7': 788, 'm8': 900, 'm9': 1047,
                  'm10': 1100, 'm11': 1227, 'm12': 1381, 'm13': 1499, 'm14': 1589}

for column, count in missing_counts.items():
    train_data[column].fillna(value=train_data[column].mean(), inplace=True)  # Replace missing values with the median of the column

# Verify if missing values have been handled
print(train_data.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Perform label encoding on 'pc' and 'ma' columns
train_data[['pc', 'ma']] = train_data[['pc', 'ma']].apply(LabelEncoder().fit_transform)
```

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split

from sklearn.model_selection import GridSearchCV

from sklearn import metrics
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.utils import class_weight
```

```python
# Split the data into features (X) and target (y)
X = train_data.drop('pred', axis=1)
y = train_data['pred']
```

```python
# Perform class balancing with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Calculate class weights
class_weights = class_weight.compute_class_weight('balanced', classes=[0, 1], y=y_train)
```

```python
# Initialize the XGBoost classifier
xgb_classifier = xgb.XGBClassifier(scale_pos_weight=class_weights[1])
```

```python
# Train the classifier on the training data
xgb_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = xgb_classifier.predict(X_test)
```

```python
y_pred
```

```python
# Calculate the F1 score
f1 = f1_score(y_test, y_pred)

# Print the F1 score
print("F1 Score:", f1)
```

```python
# Evaluate the model performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
